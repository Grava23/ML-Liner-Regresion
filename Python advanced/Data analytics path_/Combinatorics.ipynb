## 2.
Студенческая группа состоит из 23 человек, среди которых 10 юношей и 13 девушек. Сколькими способами можно выбрать двух человек одного пола?

### Сочетание

```python
import numpy as np

#Функция для расчёта сочетания
def combination(n,m):
    n_f = np.math.factorial(n)
    m_f = np.math.factorial(m)
    return n_f / (m_f*np.math.factorial(n - m))
```

```python
boys = combination(10,2)
girls = combination(13,2)
print(int(boys + girls))
```

В кошельке находится достаточно большое количество 1-, 2-, 5- и 10-рублёвых монет. Сколькими способами можно извлечь три монеты из кошелька (монеты одинакового достоинства неразличимы)?


### Сочетание с повторениями

```python
def combination_with_repetitions(n,m):
    n_f = np.math.factorial(n)
    m_f = np.math.factorial(m)
    return np.math.factorial(n + m - 1) / (m_f * np.math.factorial(n - 1))
```

## 24.
У комбинации ООР (первого игрока) комбинация  РРР (второго игрока):

```python
import random
alice = 0
bill = 0
bill_total = 0
alice_total = 0
for i in range(10000):
    game = 1
    sequence = 0
    length = 0
    while game:
        coin = random.randint(1, 2) # выбрасыем Орел = 1 или Решка = 2
        sequence = (sequence%100)*10 + coin # сдвигаем окно 3 последних бросков, 112 + 1 => 121
        if sequence == 112: # ООР
            alice += length
            alice_total += 1
            game = 0
        if sequence == 222: # РРР
            bill += length
            bill_total += 1
            game = 0
        length += 1
        
first = bill/bill_total
second = alice/alice_total
percent = first/second * 100
print ("Первый игрок:", bill/bill_total, "Второй игрок:", alice/alice_total)
print("Комбинации второго выиграет с вероятностью ",int(percent),"%")
```

```python
#подключаем нужные нам библиотеки
#вывод matplotlib прямо в окне браузера
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
#получаем данные по указанной ссылки, в качестве разделителя указываем ";"
data = pd.read_csv("https://video.ittensive.com/python-advanced/data-9722-2019-10-14.utf.csv", delimiter=";")
#перобразуме данные по районам (чтобы устранить дубликаты и сделать подписис короче),
# удаляем из название слово "район", дополнительно назначаем район, категорией,
# чтобы группировка происходила быстрее 
data["District"] = data["District"].str.replace("район ","").astype("category")
#преобразуем данные по округам, удаляем все слова кроме перового, данные по округам также преобразуем в категорию
data["AdmArea"] = data["AdmArea"].apply(lambda x:x.split(" ")[0]).astype("category")
#установим индекс по году, выполним сортировку,
#чтобы получить данные только за 2018-2019 год, затем сбросим индекс
data = data.set_index("YEAR").loc["2018-2019"].reset_index()

#поскольку данных по округам и районам достаточно много, и вместе они образуют совокупность
#будем использовать круговую диаграмму
#выведем две круговые диаграммы в одну строку
#берём холст 12 на 12
fig = plt.figure(figsize=(12,12))
#добовляем на него первую область, 2 столбца, 1 строкуй
area = fig.add_subplot(1, 2, 1)
#строим диаграмму разпределения отличников по округам, добавим заголовок данных "ЕГЭ в Москве"
#размер шрифта 20 пикселей
area.set_title("ЕГЭ в Москве", fontsize=20)
#в качестве индекса, выбираем серию данных "AdmArea" (административные округа)
data_adm = data.set_index("AdmArea")
#теперь отсечём данные только отличников, сгруппируем данные по округам, выведем сумму на ркговой диаграмме
data_adm["PASSES_OVER_220"].groupby("AdmArea").sum().plot.pie(ax=area, label="")

#для второй диаграммы выберем все данные в Северозападном округе
#добавим область для вывода данных, 
area = fig.add_subplot(1, 2, 2)
#в эту область выведем заголовок: "ЕГЭ в Северо-Западном администротивном округе"
area.set_title("ЕГЭ в СЗАО", fontsize=20)
#получаем данные по Северо-Западному административному округу, сбрасываем индекс, 
#затем устанавливаем индекс по серии данных "District", то есть по районам
data_district = data_adm.loc["Северо-Западный"].reset_index().set_index("District")
#после этого отсечём только отличников, сгруппируем по серии данных "District",
# выполним операцию суммирования
data_district = data_district["PASSES_OVER_220"].groupby("District").sum()
#вычислим общее число отличников по районам
total = sum(data_district)
#построим круговую диаграмму по районам, округляем значения отличников, а затем приводим их к целоом числу
data_district.plot.pie(ax=area, label="", autopct=lambda x:int(round(total * x/100)))
plt.show()
```
